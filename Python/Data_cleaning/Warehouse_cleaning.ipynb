```python
is_non_numeric = pd.to_numeric(df['Quantity'],errors = 'coerce').isna()  
print(is_non_numeric)

df.head()
```

```python
word_so_map = {'two hundred': 200}

```

```python
df['Quantity'] = df['Quantity'].replace(word_so_map, regex = False)
```

```python
df['Quantity'] = df['Quantity'].dropna()
```

```python
df = df.dropna(subset = ['Quantity'])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# converting status into 0's and 1's as it will be easy for further analyis and for ML 

# that procedure is called as encoding 

status_mapping = { 'In Stock': 1, 'Out of Stock': 0 }
#whats happening here is that using mapping we are telling that instock is 1 and outofstock is 0 
#now we apply it as earlier it was just assigned thats it 

df['Status'] = df['Status'].map(status_mapping)
#now its applied 
#now we will print it and see 

df['Status']
#here we have a issue that is nan and 1.0 and 0.0 , as i want only 0 and 1 so 

# we do astype to convert it to 0 and 1 in the next one 
```

```python
#here below code is for converting nan values to 0 as it has no use and 0 and nan is same so we did it 
df['Status'] = df['Status'].fillna(0)
#after that now we are converting to astype for getting 0 and 1 
df['Status'] = df['Status'].astype(int)
df['Status']
#perfect !!
```

```python
#now we will check fully once what else is there to do 

```

```python
#now coming to price 
#we will check for unique values apart from numbers 
Unique_Price = df['Price'].unique()
print(Unique_Price)
#ok !


```

```python
#now we will turn nan values to 0 
df['Price']= df['Price'].fillna(0)
df['Price']
```

```python
#now what i see is that the price is in decimals so we either round off or consider floor or ground values 
#lets see how to do that 
df['Price']= df['Price'].round(0)
df['Price']
```

```python
df['Price']= df['Price'].astype(int)
df['Price']

#perfect !! 
```

```python
df['Last Restocked'] = df['Last Restocked'].fillna(0)
df['Last Restocked']
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
cleaned_df = df


print("Saving cleaned data to cleaned_data.csv...")
cleaned_df.to_csv('cleaned_data.csv', index=False)
print("Save complete.")
```

```python

category_qty = df.groupby("Category")["Quantity"].sum()

plt.figure()
plt.bar(category_qty.index, category_qty.values)
plt.xlabel("Category")
plt.ylabel("Total Quantity")
plt.title("Total Quantity per Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.hist(df["Price"])
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Product Prices")
plt.tight_layout()
plt.show()
```

```python
warehouse_counts = df["Warehouse"].value_counts()

plt.figure()
plt.pie(warehouse_counts, labels=warehouse_counts.index, autopct="%1.1f%%")
plt.title("Product Share by Warehouse")
plt.tight_layout()
plt.show()
```
